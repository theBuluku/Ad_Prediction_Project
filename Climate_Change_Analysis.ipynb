```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
def read_dataframe(filename):
    """
    python function to load the world bank data.
    args:
        filename - path of file to be loaded
    Returns:
        normal dataframe and a transposed dataframe on country name
    """
    normal_df = pd.read_csv(filename, skiprows=4)
    t_df = normal_df.set_index('Country Name').T
    return normal_df, t_df

normal_df, t_df = read_dataframe('world_bank_data.csv')
normal_df.head()
```
